```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
#1.สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน 

x = np.arange(0,20)
y = x**2
plt.plot(x,y)
plt.title('First Time With Matplotlib')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#2.สร้างกราฟฟังก์ชั่น Sin (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสี
#เส้นเป็นสีชมพู

x = np.arange(0,100)
y = np.sin(x)
plt.plot(x,y, c = 'm')
plt.title('Sin Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#3.สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสี
#เส้นเป็นสีเขียวอ่อนโดยใช้ค่ารหัสสีฐานสิบหก

y = np.arange(0,100)
x = y**2
plt.plot(x,y, c = '#CCFF66')
plt.title('Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#4.สร้างกราฟฟังก์ชั่น Cos (เหมือนตัวอย่าง) พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน เปลี่ยนสี
#เส้นเป็นสีแดง ใส่ Line Style เป็นเส้น - -

x = np.arange(0,100)
y = np.cos(x)
plt.plot(x,y, c = 'r', ls = '--')
plt.title('Graph')
plt.xlabel('x')
plt.xlabel('y')
```

```python
#5.สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ Y = X^3 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อ
#กราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker
#และเส้นสองใช้สามเหลี่ยมเป็น marker

x = np.arange(0,20)
y1 = x**2
y2 = x**3

plt.plot(x,y1, c = 'r', marker = 'o')
plt.plot(x,y2, c = 'b', marker = '^')
plt.title('Graph')
plt.xlabel('x')
plt.xlabel('y')
```

```python
#6. สร้างกราฟธรรมดาเป็นฟังก์ชั่น Y = X^2 และ X = Y^2 พล็อตลงในรูปเดียวกัน พร้อมใส่ชื่อ
#กราฟ รายละเอียดแต่ละแกน ใช้สีแดงและน้ำเงินตามลำดับ โดยเส้นแรกใช้วงกลมเป็น marker
#และเส้นสองใช้สามเหลี่ยมเป็น marker และใส่ legend และเพิ่มเส้นตารางข้างหลัง

x = np.arange(0,20)
y = x**2
x2 = y**2

plt.plot(x,y, c = 'r', marker = 'o', label = 'y = x**2 line')
plt.plot(x2,y, c = 'b', marker = '^', label = 'x = y**2 line')
plt.title('Graph')
plt.legend(loc=0)
plt.xlabel('x')
plt.xlabel('y')
```

```python
#7. สร้างกราฟธรรมดาเป็นฟังก์ชั่น X = Y^2 + 4*Y พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน
#เปลี่ยนสีเส้นเป็นสีเหลืองอ่อนโดยใช้ค่ารหัสสีฐานสิบหก จำกัดแกน X และ Y ให้เหมาะสม

y = np.arange(0,100)
x = y**2+4*y

plt.plot(x,y, c = '#FFFF00')
plt.title('Graph')
plt.xlabel('x')
plt.ylabel('y')
```

```python
#8. สร้างกราฟใหญ่ด้วยสมการ X = Y^3 และกราฟเล็กด้วย Y = X^2 + 2 พล็อตลงในรูปเดียวกัน
#พร้อมใส่ชื่อกราฟ รายละเอียดแต่ละแกน และใส่ legend

y = np.arange(0,100)
x = y**3
y2 = x**2+2
fig = plt.figure(figsize=[7,5])
ax1 = fig.add_axes([0.1,0.1,1,1])
ax1.plot(x,y)
ax2 = fig.add_axes([0.65,0.15,0.4,0.4])
ax2.plot(x,y2)
```

```python
#9. ให้หาอัตราส่วนตั๋วโดยสารชั้น 1 2 3 และพล็อตเป็น Pie chart

df = pd.read_csv('train.csv')

Group_Class = df.groupby(['Pclass']).count()['PassengerId']

plt.figure(figsize=[5,5])
labelsClass = ['1st class', '2nd class', '3rd class' ]
plt.pie(Group_Class,labels = labelsClass)
plt.show()
```

```python
#10. ให้หาอัตราส่วนของจุดหมายปลายทาง (Embark) และพล็อต Pie chat พร้อมทั้งให้
#Queenstown เป็น explode

Group_Embark = df.groupby(['Embarked']).count()['PassengerId']

plt.figure(figsize = [5,5])
exEmbark = [0,0.1,0]
labelsEmbark = ['Cherbourg', 'Queenstown', 'Southampton']
plt.pie(Group_Embark, labels = labelsEmbark, explode = exEmbark)
plt.show()
```

```python
#11. นำราคาตั๋วโดยสารทั้งหมดมาหา histogram

Group_Fare = df.groupby(['Fare']).count()['PassengerId']
plt.figure(figsize = [8,5])
plt.hist(Group_Fare, bins = 30)
plt.show()
```

```python
#12.สร้างพล็อตแบบกระจายโดยให้อายุเป็นแกน X และ ราคาตั๋วโดยสารเป็นแกน Y

Age = df['Age']
Fare = df['Fare']

plt.figure(figsize = [9,7])
plt.scatter(Age,Fare, c = "r", linewidth=0.1)
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
#13.สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารเฉลี่ยแต่ละชั้น

Pclass_Mean = df.groupby(['Pclass']).mean()['Fare']
Idex_Pclass = np.array(Pclass_Mean.index)
plt.figure(figsize = [5,5])
plt.bar(Idex_Pclass, Pclass_Mean)
plt.show()
```

```python
#14. สร้าง bar chart โดยใช้ข้อมูลจากราคาตั๋วโดยสารต่ออายุของผู้โดยสาร

plt.figure(figsize = [5,5])
plt.bar(Age, Fare)
plt.show()
```

```python
#15. สร้าง bar chart โดยใช้ข้อมูลผู้ที่นามสกุลซ้ำกัน โดยความถี่เป็นแกน Y
df10 = pd.read_csv('train.csv')
df10['LastName'] = df10['Name'].str.split(',', expand=True)[0]
df11 = df10[df10['LastName'].duplicated()]
df12 = df11.groupby(['LastName']).size()

plt.figure(figsize = [50,5])
plt.bar(df12.index, df12.values)
plt.xticks(rotation=90)
plt.show()
```
